```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('911.csv')
```

```python
'''
v1:
def split(x):
    return x.split(':')[0]
v2:
df['Reason'] = df['title'].apply(split)
v3:
df['Reason'] = df['title'].map(lambda x: x.split(':')[0])
'''

df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

```python
df['Reason'].value_counts().idxmax()
```

```python
sns.set_style('darkgrid')
sns.countplot(data=df, x='Reason', hue='Reason')
```

```python
df['timeStamp'].info(verbose=True)
```

```python
df['timeStamp']= pd.to_datetime(df['timeStamp']) 
# df['timeStamp'].apply(pd.to_datetime)
df['timeStamp'].info(verbose=True)
```

```python
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
# df['Day of Week'] = df['timeStamp'].apply(lambda x: x.day_name())
df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
ax = sns.countplot(data=df, x='Day of Week', hue='Reason', palette='viridis') # v1
sns.move_legend(ax, 'center right', bbox_to_anchor=(1.2,0.5))
# plt.legend(loc='center right', bbox_to_anchor=(1.2, 0.5))
```

```python
ax = sns.countplot(data=df, x='Month', hue='Reason', palette='viridis') # v1
sns.move_legend(ax, 'center right', bbox_to_anchor=(1.2,0.5))
```

```python
dfMonth = df.groupby(['Month']).count()
dfMonth.head()
```

```python
# sns.lineplot(data=dfMonth, x='Month', y='lat')
dfMonth['timeStamp'].plot()
```

```python
sns.lmplot(data=dfMonth.reset_index(), x='Month', y='timeStamp') # df with the new index
```

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
df.head()
```

```python
df.groupby(['Date']).count()['timeStamp'].plot()
plt.title('Date vs Calls')
```

```python
dfReason = df.groupby(['Reason']).count()

dfReason.head()
```

```python
df[df['Reason'] == 'EMS'].groupby('Date').count()['timeStamp'].plot()
plt.title('Date vs Calls: Reason - EMS')
```

```python
df[df['Reason'] == 'Fire'].groupby('Date').count()['timeStamp'].plot()
plt.title('Date vs Calls: Reason - Fire')
```

```python
df[df['Reason'] == 'Traffic'].groupby('Date').count()['timeStamp'].plot()
plt.title('Date vs Calls: Reason - Traffic')
```
